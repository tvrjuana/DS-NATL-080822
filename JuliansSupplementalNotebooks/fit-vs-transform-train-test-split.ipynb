```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Data generated by me flailing at my keyboard

raw_data = pd.DataFrame([1,2,5,7,8,9,1,2,4,6,7,8,10,4,6,2,5,7,14,15,13,12,11,11,11,16,17,14,3,5], columns = ['Raw'])
```

```python
raw_data
```

```python
ss1 = StandardScaler()

allscale = pd.DataFrame(ss1.fit_transform(raw_data), index = raw_data.index, columns = ['all_ss1'])
allscale
```

```python
unscaledresults = pd.DataFrame(ss1.inverse_transform(allscale), index = allscale.index, columns = ['all_ss1_invert'])
unscaledresults
```

```python
resultsdf = pd.concat([raw_data, allscale, unscaledresults], axis = 1)

resultsdf
```

```python
check = resultsdf.Raw - resultsdf.all_ss1_invert

#These should all be zero
check
```

```python
#hypothetical train test split

#setting random seed for repeatability
np.random.seed(14)
#This selects 5 random intergers between 0 and 29, which we can map to the index of our underlyingdata
test_data_index = np.random.choice(range(30), size=5, replace=False)
```

```python
test_data_index
```

```python
#The index values for our test data willl just be our 5 chosen numbers
test_data = raw_data.iloc[test_data_index]
display(test_data)

# Index values for the training data will be 0-29 minus those numbers.

train_data_index = list(range(30))

for i in test_data_index:
    train_data_index.remove(i)
train_data = raw_data.iloc[train_data_index]

display(train_data)
```

```python
resultsdf = pd.concat([resultsdf, trainscaleresults, trainscaleinvert], axis = 1)
```

So far so good. But! Note the difference in our all_ss1 and the train_sstrain.

But, you may say, if I apply all steps to the test data that I applied to the training data, I should be fine. Let's look at that w/r/t our scaler:

```python
# Here I have copy and pasted all my code from above, and will simply put in 'test' everwhere there is a 'train'

# sstrain = StandardScaler()
sstest = StandardScaler()

# trainscaleresults = pd.DataFrame(sstrain.fit_transform(train_data), index = train_data.index, columns = ['train_sstrain'])
testscaleresults = pd.DataFrame(sstest.fit_transform(test_data), index = test_data.index, columns = ['test_sstest'])

# trainscaleinvert = pd.DataFrame(sstrain.inverse_transform(trainscaleresults), index = trainscaleresults.index, columns = ['train_sstrain_invert'])
testscaleinvert = pd.DataFrame(sstest.inverse_transform(testscaleresults), index = testscaleresults.index, columns = ['test_ssttest_invert'])

# trainscaleresults
testscaleresults
```

Looks fine and normal, but what if we tried to put this into a model fit on the training data?



```python
# In the training data, the 'underlying' values associated with the values of test_sstest is: 
newxtest = sstrain.inverse_transform(testscaleresults)

newxtest
```

```python
# Compare that to the true values:

test_data
```

So if we fed this into our model, our predictions would be given out for the wrong data points. The model would be trying to give us the asnwer for an 'underlying' data point of approx 4.5, 12, 2, 6.2, 13.7 and so on. The predictions would be true to the model AT THOSE POINTS. But we'd be ascribing them to the underlying points of 5, 14, 2, 7, and 16.  This would, of course, generate inexplicably inaccurate results. 

Let's try to do this properly.

```python
# Copy and pasted from above:

# sstrain = StandardScaler()
# sstest = StandardScaler()
# I'm commenting out this option entirely, we *don't* want to generate a new scaler 


# trainscaleresults = pd.DataFrame(sstrain.fit_transform(train_data), index = train_data.index, columns = ['train_sstrain'])
# testscaleresults = pd.DataFrame(sstest.fit_transform(test_data), index = test_data.index, columns = ['test_sstest'])
# Here we'll keep most things as above, but we will only transform using the training scaler. I'll also adjust the column name. 

testscaleresults = pd.DataFrame(sstrain.transform(test_data), index = test_data.index, columns = ['test_sstrain'])

# trainscaleinvert = pd.DataFrame(sstrain.inverse_transform(trainscaleresults), index = trainscaleresults.index, columns = ['train_sstrain_invert'])
testscaleinvert = pd.DataFrame(sstrain.inverse_transform(testscaleresults), index = testscaleresults.index, columns = ['test_ssttest_invert'])

# trainscaleresults
display(testscaleresults)
display(testscaleinvert)
```

### Okay, let's see this in action with a model.

For this dataset, let's assume we're a restaurant, catering exclusively to couples, and we want to calculate how many chairs we'll need on a night, given the number of couples who have booked a reservation. We also keep 10 tables (so 20 chairs) for walk in customers. 

I will generate the X data by randomly hitting my keyboard, and y data by taking the X, multiplying by 2, and adding 10.

```python
x = [12,14,12,23,53,73,1,3,45,32,14,24,74,86,2,5,67,34,7,2,64,435,98,2,6,234,45,14,34,8]
```

```python
len(x)
```

```python
couplesandchairs_df = pd.DataFrame(x, columns = ['reservation_couples'])
```

```python
couplesandchairs_df['chairs'] = couplesandchairs_df.reservation_couples*2 + 20

couplesandchairs_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(couplesandchairs_df.reservation_couples, couplesandchairs_df.chairs,  random_state = 14)
```

```python
display(X_train.head())
display(y_train.head())
display(X_test.head())
display(y_test.head())
```

```python
# Now let's build a model

sschairs = StandardScaler()

train_chairs_scaled = pd.DataFrame(sschairs.fit_transform(np.asarray(X_train).reshape(-1, 1)), index = X_train.index)

model = LinearRegression()

model.fit(train_chairs_scaled, y_train)
```

```python
train_chairs_scaled
```

```python
model.predict(np.asarray(X_test).reshape(-1, 1))
```

```python
X_test_scaled = pd.DataFrame(sschairs.transform(np.asarray(X_test).reshape(-1, 1)), index = X_test.index)
X_test_scaled
```

```python
y_hat = model.predict(X_test_scaled)
y_hat
```
